```python
import requests
import pandas as pd
import json
url = "https://api.met.no/weatherapi/locationforecast/2.0/compact?"
params = { 'lat':9.93
          ,'lon':76.27
        }

headers = {
    "User-Agent": "weather-ml-learning/1.0 snandhakishor583@gmail.com"}
json_data = requests.get(url, params=params, headers=headers).json()
rows = json_data['properties']['timeseries']
new_data = []
for row in rows:
  info = { 'time' : row['time'],
           'air_pressure': row['data']['instant']['details']['air_pressure_at_sea_level'],
           'temperature' : row['data']['instant']['details']['air_temperature'],
           'cloud_area_fraction' : row['data']['instant']['details']['cloud_area_fraction'],
           'relative_humidity'  : row['data']['instant']['details']['relative_humidity'],
           'wind_from_direction' : row['data']['instant']['details']['wind_from_direction'],
           'wind_speed' : row['data']['instant']['details']['wind_speed'],


       }

  new_data.append(info)

```

```python
df = pd.DataFrame(new_data)
df['time'] = pd.to_datetime(df['time'], errors="coerce")
df.head(10)
```
